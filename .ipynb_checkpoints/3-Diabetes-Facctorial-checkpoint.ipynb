```python
# Aquí empieza la fiesta !!!!
from sklearn.model_selection import train_test_split
def partirDatos(datos):
    return train_test_split( datos.drop("Outcome", axis=1) , datos["Outcome"] ,test_size = .25, random_state = 60) # semilla
```

```python
from sklearn.preprocessing import StandardScaler

def reemplazarValoresAtipicos(datos, columna):
    Q1 = datos[columna].quantile(q=0.25)
    Q3 = datos[columna].quantile(q=0.75)
    RIC = Q3-Q1
    limite_maximos = Q3 + 1.5 * RIC
    limite_minimos = Q1 - 1.5 * RIC
    datos[columna] = datos[columna].apply(lambda dato: limite_maximos if dato>limite_maximos else dato)
    datos[columna] = datos[columna].apply(lambda dato: limite_minimos if dato<limite_minimos else dato)
    return datos
    
def rellenarValoresNulos(datos, columna):
    Q2 = datos[columna].quantile(q=0.5)
    datos[columna]=datos[columna].replace(np.NaN, Q2)
    return datos
    
def transformacion(datos):
    for columna in datos.columns:
        # Paso 1: Qué pasa con los nulos?
        #   Estrategias: Les pongo la mediana... Les pongo el valor promedio de los que se parezcan a él
        datos=rellenarValoresNulos(datos, columna)
        # Paso 2: Atípicos
        datos=reemplazarValoresAtipicos(datos, columna)
        
    # Convertir los datos a valores tipificados: Resta la media y divide entre la desv. tipica
    tipificador = StandardScaler()
    return tipificador.fit_transform(datos)
```
